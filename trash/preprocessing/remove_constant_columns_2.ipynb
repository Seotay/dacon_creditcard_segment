```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
train_df = pd.read_parquet('./data/train/train.parquet')
train_df
```

```python
test_df = pd.read_parquet('./data/test/test.parquet')
test_df
```

```python
print('shape of train set in member: ', train_df.shape)
print('shape of test set in member: ', test_df.shape)
```

```python
constant_columns_train = train_df.nunique() == 1
constant_columns_train = constant_columns_train[constant_columns_train].index.tolist()

print("Extract columns to have duplication values :\n", constant_columns_train)
print(len(constant_columns_train))
```

```python
constant_columns_test = test_df.nunique() == 1
constant_columns_test = constant_columns_test[constant_columns_test].index.tolist()

print("Extract columns to have duplication values :\n", constant_columns_test)
print(len(constant_columns_test))
```

```python
union_constant_columns = list(set(constant_columns_train) | set(constant_columns_test))
print(len(union_constant_columns))
```

```python
train_df.drop(columns=union_constant_columns).to_parquet('./data/train/train_filtered.parquet', index=False)
test_df.drop(columns=union_constant_columns).to_parquet('./data/test/test_filtered.parquet', index=False)
```
